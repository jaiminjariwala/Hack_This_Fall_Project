### **Building a Popularity Based Recommendation System**
* 1. Criteria: we will show the top 50 highest average rating books
* 2. Criteria: we will consider only those books on which minimum 250 people has voted!

```python
# let's firstafall combine the 'books' dataset and 'ratings dataset' on the top of ISBN column
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

#### **Now from this dataset, we will find out average rating of every books, and will consider only those books whose having votes > 250**

```python
# let's find out how many votes are there on each book
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
# merging 'num_rating_df' and 'avg_rating_df'
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
# Now I want only those books whose num_ratings are greater than or equal to 250
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

### **Hence Popularity based Filtering Recommendation system done**

## **Now working on Collaborative Filtering Based Recommendation System**

```python
ratings_with_name
```

```python
# will keep only those users who has done minimum 200 ratings
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
intelligent_users = x[x].index
# so this will provide us a result as: Which User has rated how many books
```

```python
# get me those rows from 'ratings_with_name' those are 'intelligent_user'
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(intelligent_users)]
filtered_rating
```

### **We have done filtering based on the users, now we have to do filtering based on Books** 

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books =  y[y].index
famous_books
# every books and how many ratings have been done on it...
# this are those books on which more than 50 ratings are there
```

```python
final_ratings =  filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates() # if any !
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# now we have to carry out the cosine_similarity() of every rows with other rows i.e...
# We will calculate the EUCLIDIAN DISTANCE of every 1 vector with other vectors.

similarity_scores = cosine_similarity(pt)
similarity_scores

# we will be getting a vector, a 2-D numpy array 
```

```python
similarity_scores.shape

# so basically we have calculated 706 books Euclidian Distance with 706 books
```

```python
# let's check the 1st book's similarity score with every other books, i.e checking the euclicdian distance of 1st book with every other books
similarity_scores[0]
```

```python
# Now let's write a function, that will take a Book as an input and will return you 5 books as in recommendation

def recommend(book_name):
    # fetching the index of so and so book 
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x : x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])    # as to get the index of the most similar items
```

```python
print("5 Books Recommended based on this movie:\n")
recommend('1984')
```
